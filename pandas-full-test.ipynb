# **NULL/EMPTY Cell values**

Filling empty/null values with a given user choice input values.

```python

songs2 = songs.fillna("Unkonwn")
songs2
```

```python
# songs3 = songs2.drop([1000, 1001, 1002], axis=0)
songs3 = songs2.drop(songs2.index[1000:1003], axis=0)
songs3
```

# **Grouping, Sorting and Indexing**

```python
# Grouping values by a column
songs3.groupby('artist').artist.count()

# Setting an Index and Sorting
songs3.set_index("dur").sort_values(by='dur', ascending=False)
```

# **Renaming Columns**

```python
songs4 = songs3.rename(columns={'dur': 'play_time'})
songs4.set_index("play_time").sort_values(by='play_time', ascending=False)
```
